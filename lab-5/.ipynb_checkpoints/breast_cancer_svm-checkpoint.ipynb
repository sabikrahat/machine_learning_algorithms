# <center>Support Vector Machine</center>

## Importing Libraries

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn import svm
from sklearn import metrics
import pandas as pd
import numpy as np
import seaborn as sns
```

## load dataset from sklearn dataset module

```python
skdata = load_breast_cancer()
skdata
```


## turning skdata into a dataframe

```python
df = pd.DataFrame(skdata.data, columns=skdata.feature_names)
df
```

## adding target colum in the dataframe

```python
df['target'] = pd.Series(skdata.target)
df.head(10)
```

## creating independent and dependent dataset

```python
x = df.drop('target',axis=1)
x
```

```python
y = df['target']
y
```

## splitting data for training and testing

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

## loading SVM model

```python
clf = svm.SVC(kernel="linear")
clf
```

## fitting SVM model

```python
clf.fit(x_train, y_train)
```

## predicting the test dataset

```python
predict = clf.predict(x_test)
predict
```

## generating confusion matrix

```python
pd.DataFrame(confusion_matrix(y_test,predict),columns=['Predicted No','Predicted Yes'],index=['Actual No','Actual Yes'])
```

## generating report

```python
print(classification_report(y_test,predict))
```
